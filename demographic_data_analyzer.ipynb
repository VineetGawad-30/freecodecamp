```python
import kagglehub

# Download latest version
path = kagglehub.dataset_download("uciml/adult-census-income")

print("Path to dataset files:", path)
```

```python
import pandas as pd
import os

# Define the path from your screenshot
path = "/kaggle/input/adult-census-income"

# Find the CSV file in that folder
files = os.listdir(path)
csv_file = [f for f in files if f.endswith('.csv')][0]
full_path = os.path.join(path, csv_file)

# Load the dataset
df = pd.read_csv(full_path)

# Show the first few rows to verify
print(f"Loaded: {csv_file}")
df.head()
```

```python
rename_dict = {
    'education.num': 'education-num',
    'marital.status': 'marital-status',
    'capital.gain': 'capital-gain',
    'capital.loss': 'capital-loss',
    'hours.per.week': 'hours-per-week',
    'native.country': 'native-country',
    'income': 'salary'
}
df.rename(columns=rename_dict, inplace=True)
```

```python
race_count = df['race'].value_counts()
race_count
```

```python
#Average age of men
average_age_men = round(df[df['sex'] == 'Male']['age'].mean(), 1)
```

```python
average_age_men
```

```python
#Percentage with Bachelors
percentage_bachelors = round((df['education'] == 'Bachelors').sum() / len(df) * 100, 1)
percentage_bachelors
```

```python
higher_education = df['education'].isin(['Bachelors', 'Masters', 'Doctorate'])
lower_education = ~higher_education
```

```python
lower_education
```

```python
higher_education_rich = round(
    (df[higher_education & (df['salary'] == '>50K')].shape[0] /
     df[higher_education].shape[0]) * 100, 1
)
```

```python
higher_education_rich
```

```python
lower_education_rich = round(
    (df[lower_education & (df['salary'] == '>50K')].shape[0] /
     df[lower_education].shape[0]) * 100, 1
)
```

```python
lower_education_rich
```

```python
min_work_hours = df['hours-per-week'].min()
min_work_hours
```

```python
num_min_workers = df[df['hours-per-week'] == min_work_hours]
rich_percentage = round(
    (num_min_workers[num_min_workers['salary'] == '>50K'].shape[0] /
     len(num_min_workers)) * 100, 1
)
```

```python
rich_percentage
```

```python
country_stats = df.groupby('native-country')['salary'].value_counts(normalize=True).unstack()
highest_earning_country = country_stats['>50K'].idxmax()
highest_earning_country_percentage = round(country_stats['>50K'].max() * 100, 1)
```

```python
top_IN_occupation = df[
    (df['native-country'] == 'India') & (df['salary'] == '>50K')
]['occupation'].value_counts().idxmax()
```

```python
print(f"Average age of men: {average_age_men}")
print(f"Highest earning country: {highest_earning_country} ({highest_earning_country_percentage}%)")
print(f"Top occupation in India: {top_IN_occupation}")
```
